```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('../data/retail_case_data.csv')
```

## Plotting the chart

```python
import matplotlib.pyplot as plt

# education mapping dictionary
map_educ = {'2n Cycle': 1, 'Basic': 2, 'Graduation': 3, 'Master': 4, 'PhD': 5}

# applying the map dict to transform categorical value into quantitative variable
df['Numeric_Graduation'] = df['Education'].map(map_educ)



# plotting the chart
true_frequeny_by_education = df.groupby('Education')['Response'].mean()
true_frequeny_by_education.plot(kind='bar', figsize=(15,6))
plt.title('True frequency by Education')
plt.xlabel('Education')
plt.ylabel('True frequency')

plt.show()
```

## Viewing the correlation matrix

```python
to_corr = df.groupby('Numeric_Graduation')['Response'].mean().reset_index()

corr_matrix = to_corr.corr()


print("Matriz de Correlação:")
print(corr_matrix)
```

## Conclusion

As we can see, there is a strong relationship between graduation and response rate, where the correlation index is 0.795867 indicating that the majority of the customers who's accepted the offer have higher education level.

This can lead us to believe that the higher customer's education level is, higher is the probability to him accepts the offer.
